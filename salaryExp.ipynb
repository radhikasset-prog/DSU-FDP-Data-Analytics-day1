```python
import matplotlib.pyplot as plt
import numpy as np

# Generate synthetic data for experience and salary
np.random.seed(42) # for reproducibility
experience = np.random.uniform(1, 20, 50) # 50 data points, experience from 1 to 20 years
salary = 30000 + experience * 5000 + np.random.normal(0, 8000, 50) # Base salary + 5000 per year + some noise

# Create the plot
plt.figure(figsize=(10, 6))
plt.scatter(experience, salary, color='blue', alpha=0.7)

# Add labels and title
plt.title('Salary vs. Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary ($)')
plt.grid(True)
plt.show()
```
