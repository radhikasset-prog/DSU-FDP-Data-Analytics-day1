```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("student_result.csv")
print(df)
max_cat1=df["CAT1"].max()
print("Max-CAT1",max_cat1)
min_cat1=df["CAT1"].min()
print("Min of CAT1",min_cat1)

plt.figure(figsize=(18, 6)) # Adjust the  figure size

plt.subplot(1,2,1) #single space is divided into 2 ,and shows in 1st portion
plt.bar(df["Student"],df["CAT1"],color="green")
plt.title("CAT1 Performance")
plt.legend()
plt.xlabel("Student") #  x-axis label
plt.ylabel("CAT1 Score") #  y-axis label for
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability

plt.subplot(1,2,2) #single space is divided into 2 ,and shows in 2nd portion
plt.bar(df["Student"],df["CAT2"],color="blue")
plt.title("CAT2 Performance") # Added title for the second subplot
plt.xlabel("Student") #  x-axis label
plt.ylabel("CAT2 Score") #  y-axis label
plt.xticks(rotation=30, ha='right') # Rotate x-axis labels for better readability

#plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```
